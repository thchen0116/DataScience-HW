# crawler THSR information
Let's start to crawl the imformation of each station of THSR!
* Target web: [THSR Homepage](https://www.thsrc.com.tw)  
* Choose each station:
    1. Homepage -> Travel Information -> Stations
    2. choose different stations by changing url
* Task:
    1. crawler the name, address, operation hours and ticketing hours of each station
    2. organize into a DataFrame

```python
# setting each station name and their url
a=requests.get("https://www.thsrc.com.tw/ArticleContent/2f940836-cedc-41ef-8e28-c2336ac8fe68")
b=requests.get("https://www.thsrc.com.tw/ArticleContent/977abb69-413a-4ccf-a109-0272c24fd490")
c=requests.get("https://www.thsrc.com.tw/ArticleContent/e6e26e66-7dc1-458f-b2f3-71ce65fdc95f")
d=requests.get("https://www.thsrc.com.tw/ArticleContent/fbd828d8-b1da-4b06-a3bd-680cdca4d2cd")
e=requests.get("https://www.thsrc.com.tw/ArticleContent/a7a04c89-900b-4798-95a3-c01c455622f4")
f=requests.get("https://www.thsrc.com.tw/ArticleContent/e8fc2123-2aaf-46ff-ad79-51d4002a1ef3")
g=requests.get("https://www.thsrc.com.tw/ArticleContent/3301e395-46b8-47aa-aa37-139e15708779")
h=requests.get("https://www.thsrc.com.tw/ArticleContent/38b8c40b-aef0-4d66-b257-da96ec51620e")
i=requests.get("https://www.thsrc.com.tw/ArticleContent/5f4c7bb0-c676-4e39-8d3c-f12fc188ee5f")
j=requests.get("https://www.thsrc.com.tw/ArticleContent/60831846-f0e4-47f6-9b5b-46323ebdcef7")
k=requests.get("https://www.thsrc.com.tw/ArticleContent/9c5ac6ca-ec89-48f8-aab0-41b738cb1814")
l=requests.get("https://www.thsrc.com.tw/ArticleContent/f2519629-5973-4d08-913b-479cce78a356")
#抓每個車站的網址
```

```python
# crawler the address of each station
#設一個dictionary
Data = {'address':[],'operation_hours':[],'ticketing_hours':[]}
station=[a,b,c,d,e,f,g,h,i,j,k,l]
for m in station:
    soup=BeautifulSoup(m.text,"html.parser")
    string = soup.find(class_="google-map-link orange")
    Data["address"].append(string.text)
```

```python
# crawler the operation hours and ticketing hours of each station
for m in station:
    soup = BeautifulSoup(m.text,"html.parser")
    list1 = []
    string = soup.find(class_="table")
    times = string.find_all(class_="gray")
    for time in times:
        list1.append(time.text)
    Data["operation_hours"].append(list1[0].split("：")[1])
    Data["ticketing_hours"].append(list1[1].split("：")[1])
```

```python
# create a DataFrame
Data = pd.DataFrame(Data)
station = pd.DataFrame({'station':["南港","台北","板橋","桃園","新竹","苗栗","台中","彰化","雲林","嘉義","台南","左營"]})
station.join(Data)
```

# Crawler GDP and CPI
Let's start to crawl GDP and CPI!
* Target web: 
    1. [Wiki GDP](https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal))  
    2. [Wiki CPI](https://en.wikipedia.org/wiki/Corruption_Perceptions_Index)
* Task:
    1. crawler GDP table (top 50: United States ~ New Zealand)
    2. organize into a DataFrame (columns: Country, Region, IMF_Estimate, IMF_Year, United_Nations_Estimate, United_Nations_Year, World_Bank_Estimate, World_Bank_Year)
    3. crawler CPI table which contains country and 2020 CPI (top 100: Denmark ~ Suriname)
    4. organize into a DataFrame (columns: Country, CPI_2020)
    5. merge GDP(DataFrame) and CPI(DataFrame), based on Country of GDP
    6. plot and text the names of GDP top 10 countrys

```python
# crawler GDP
gdp = requests.get("https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)")
soup = BeautifulSoup(gdp.text,"html.parser")
tables = soup.find_all('tbody')
a = tables[2].find_all('tr', recursive=False)
res = []
for i in range(len(a)):
    if i < 2 or i > 51:
        continue
    tmp = a[i].text.split('\n')
    t = []
    for j in range(len(tmp)):
        #國家前面\xa0，把他刪掉，不然會造成後面判斷錯誤
        tmp[j] = tmp[j].replace(u'\xa0', '')
        if j >= 1 and j < len(tmp) - 1:
            tmpl = tmp[j].split(']')
            if len(tmpl) == 2:
                t.append(tmpl[1].replace(',', ''))
            else :
                t.append(tmpl[0].replace(',', ''))
    #存成一個list
    res.append(t)
for i in range(len(res)):
    for j in range(len(res[i])):
        if j>1:
            #把原本是字串型態的數字改成整數型態
            res[i][j]=int(res[i][j])
```

```python
# create GDP DataFrame
colname=['Country','Region','IMF_Estimate','IMF_Year','United_Nations_Estimate','United_Nations_Year','World_Bank_Estimate','World_Bank_Year']
GDP = pd.DataFrame(res, columns = colname)
GDP
```

```python
# crawler CPI
cpi = requests.get("https://en.wikipedia.org/wiki/Corruption_Perceptions_Index")
soup = BeautifulSoup(cpi.text,"html.parser")
tables = soup.find_all('tbody')
#找tbody之後發現所要的資料在第四個位置
tr = tables[4].find_all('tr')
CPI={'Country':[],"CPI_2020":[]}
for i in range(len(tr)):
    #要找第1-100筆資料
    if i<2 or i >101:
        continue
    else:
        CPI['Country'].append(tr[i].text.split("\n")[2].replace(u'\xa0', ''))
        CPI["CPI_2020"].append(int(tr[i].text.split("\n")[3]))
```

```python
# CPI DataFrame
CPI=pd.DataFrame(CPI)
CPI
```

```python
# merge GDP and CPI DataFrame
#用country去merge
Data = GDP.merge(CPI, how='inner', on='Country')
Data
```
