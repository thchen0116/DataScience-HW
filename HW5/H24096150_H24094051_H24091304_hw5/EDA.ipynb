```python
#import package
from sklearn.neural_network import MLPClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load data
data=pd.read_csv("train.csv")
X=data.iloc[:,0:-1]
y=data.iloc[:,-1]
train_X,test_X,train_y,test_y=train_test_split(X,y,train_size=0.8,stratify=y,random_state=1)
```

```python
#利用matplotlib和seaborn對特徵進行資料視覺化，觀察各特徵與y(Exited)之間的關係
fig, axarr = plt.subplots(2, 2, figsize=(20, 12))
sns.countplot(x='Geography', hue = train_y,data = train_X, ax=axarr[0][0])
sns.countplot(x='Gender', hue = train_y,data = train_X, ax=axarr[0][1])
sns.countplot(x='HasCrCard', hue = train_y,data = train_X, ax=axarr[1][0])
sns.countplot(x='IsActiveMember', hue = train_y,data = train_X, ax=axarr[1][1])
```

```python
# 利用相關性矩陣觀察各特徵之間的關聯
corr=train_X[['Age','Balance','CreditScore','EstimatedSalary','Tenure','NumOfProducts']].corr()
plt.figure(figsize=(12,12))
sns.heatmap(corr, square=True, annot=True, cmap="RdBu_r")
```
