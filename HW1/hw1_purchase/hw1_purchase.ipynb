## Purchasing Analysis (Gender)

* Run basic calculations to obtain below values by gender
  * Purchase Count
  * Average Purchase Price  
  * Total Purchase Value
  * Average Purchase Total per Person by Gender
* Create a summary data frame to hold the results
* Display the summary data frame

```python
p=purchase_data.groupby(["Gender","Purchase ID"]).mean().reset_index()
p1=p.groupby("Gender").count().reset_index()
p2=p.groupby("Gender").mean().reset_index()
p1=p1[["Gender","Purchase ID"]]
p2=p2[["Price"]].round(5)
p3=p1.join(p2)
p3=p3.rename(columns={"Purchease ID":"Purchase count","Price":"Average Purchase Price"})
p3["Total Purchase Value"]=(p3["Purchase ID"]*p3["Average Purchase Price"]).round(2)
q=purchase_data.groupby(["SN","Gender"]).count().reset_index()
q1=q.groupby("Gender").count().reset_index()
q2=q1[["Gender","SN"]]
p3["Avg Purchase Total per Person"]=(p3["Total Purchase Value"]/q2["SN"]).round(2)
a=[]
for i in range(len(p3)):
    a.append("$"+str(p3["Avg Purchase Total per Person"][i]))
p3.drop(columns="Avg Purchase Total per Person")
p3["Avg Purchase Total Person"]=(a)
p3.set_index("Gender")
```

## Top Spenders

* Identify the the top 5 spenders in the game by total purchase value (top-5 players with highest total purchase value), then list (in a table):
  * SN(screen name)
  * Purchase Count
  * Average Purchase Price
  * Total Purchase Value
* Run basic calculations to obtain the results in the table below
* Create a summary data frame to hold the results
* Sort the total purchase value column in descending order
* Display a preview of the summary data frame

```python
p=purchase_data.groupby("SN").sum().reset_index()
p1=p.sort_values(by="Price",ascending=False).head().rename(columns={"Price":"Total Purchase Value"})
p1=p1.set_index('SN')
p1=p1[["Total Purchase Value"]]
q=purchase_data.groupby("SN").count().reset_index()
q1=q.set_index("SN")
q1=q1[["Purchase ID"]].rename(columns={"Purchase ID":"Purchase Count"})
p2=q1.merge(p1, left_index=True, right_index=True)
p2=p2.sort_values(by="Total Purchase Value",ascending=False)
p2.insert(1,"Average Purchase Price",(p2["Total Purchase Value"]/p2['Purchase Count']).round(2))
a,b=[],[]
for i in range(len(p2)):
    a.append("$"+str('%.2f'%(p2["Average Purchase Price"][i]).round(2)))
    b.append("$"+str('%.2f'%(p2["Total Purchase Value"][i]).round(2)))
p3=p2.drop(columns=["Average Purchase Price","Total Purchase Value"])
p3["Average Purchase Price"]=(a)
p3["Total Purchase Value"]=(b)
p3
```

用groupby"SN"把每個人分類，用價格排序，然後重新命名，
用新的一個df計算每個人的數量，重新命名，然後把兩個資料結合，再去計算平均價格，再用for loop把價格前面都加上$，

## Most Popular Items

* Top 5 most popular items by purchase count:
  * Item ID

  * Purchase Count
  * Item Price
  * Total Purchase Value
* Retrieve the Item ID, Item Name, and Item Price columns
* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
* Create a summary data frame to hold the results
* Sort the purchase count column in descending order
* Display a preview of the summary data frame

```python
p=purchase_data.groupby(["Item ID","Item Name"]).count().reset_index()
p=p.sort_values(by="Purchase ID",ascending=False).head()
p=p[["Item ID","Item Name","Purchase ID"]].rename(columns={"Purchase ID":"Purchase Count"})
p1=purchase_data.groupby(["Item ID","Item Name"]).mean().reset_index()
p2=p.merge(p1,how='inner').sort_values(by="Purchase Count",ascending=False)
p2=p2[["Item ID","Item Name","Purchase Count","Price"]]
a,b=[],[]
for i in range(len(p2)):
    a.append("$"+str('%.2f'%((p2["Price"][i]).round(2))))
    b.append("$"+str((p2["Price"][i]*p2["Purchase Count"][i]).round(2)))
p2["Item Price"]=a
p2["Total Purchase Value"]=b
p2=p2.drop(columns="Price").set_index("Item ID")
p2
```
